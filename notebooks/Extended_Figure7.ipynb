```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = '../SourceData/Extended_Figure7/'
```

# Traces

```python
trace_data = np.load(data_path + 'a_filtered_traces.npz')
```

```python
traces = trace_data['traces']
sampling_frequency = trace_data['sampling_frequency']
active_channel_names = trace_data['active_channel_names']

plt.figure(figsize=(4, 5))

gap = 100
additional_gap = 200  

for trace_index, trace in enumerate(traces):
    if trace_index == 32:
        current_offset = trace_index * gap + additional_gap
    elif trace_index > 32:
        current_offset = trace_index * gap + additional_gap
    else:
        current_offset = trace_index * gap

    plt.plot(trace + current_offset)
    
plt.tight_layout()
plt.show()
```

```python
zoom_in_traces = traces[:, int(sampling_frequency * 0) : int(sampling_frequency * 3.5)]
plt.figure(figsize=(4, 5))

gap = 100
additional_gap = 200  

for trace_index, trace in enumerate(zoom_in_traces):
    if trace_index == 32:
        current_offset = trace_index * gap + additional_gap
    elif trace_index > 32:
        current_offset = trace_index * gap + additional_gap
    else:
        current_offset = trace_index * gap

    plt.plot(trace + current_offset)
    
plt.tight_layout()
plt.show()
```

# Beat delays

```python
beat_data = pd.read_csv(data_path + 'c_beat_delays.csv')
```

```python
plt.figure(figsize=(6,4))
plt.plot(beat_data["within_CMEC_delay"], 'o-', label="Within CM/EC")
plt.plot(beat_data["within_CM_delay"], 'o-', label="Within CM")
plt.plot(beat_data["cross_delay"], 'o-', label="Cross (CM/ECCM)")
plt.xlabel("Beat number")
plt.ylabel("Delay (s)")

plt.legend()
plt.tight_layout()
plt.show()
```

# IBI

```python
ibi_data = pd.read_csv(data_path + 'd_ibi.csv')
```

```python
plt.figure(figsize=(6,4))
plt.plot(ibi_data["ibi_cmec"], 'o-', label='CM/EC', color='steelblue')
plt.plot(ibi_data["ibi_cm"], 'o-', label='CM', color='darkorange')
plt.xlabel("Beat number")
plt.ylabel("Inter-beat interval (s)")
plt.title("Beat-to-beat IBI comparison")
plt.legend()

plt.tight_layout()
plt.show()
```
